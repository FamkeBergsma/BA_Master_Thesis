```python
import pandas as pd
import matplotlib.pyplot as plt
```

# ESG

# EU data

```python
EU_ESG = pd.read_csv(r"Cfile")

print(EU_ESG.head())
```

```python
# Count unique countries in the dataset
num_countries = EU_ESG["Country of Exchange"].nunique()

# Count the number of companies per country
country_counts = EU_ESG["Country of Exchange"].value_counts().reset_index()
country_counts.columns = ["Country", "Number of Companies"]

# Display the country count table
print(f"Number of unique countries: {num_countries}")
print(country_counts)

# Count unique industries in the dataset
num_industries = EU_ESG["NAICS National Industry Name"].nunique()

# Count the number of companies per industry
industry_counts = EU_ESG["NAICS National Industry Name"].value_counts().reset_index()
industry_counts.columns = ["Industry", "Number of Companies"]

# Display the industry count table
print(f"Number of unique industries: {num_industries}")
print(industry_counts)
```

```python
# Select only ESG score columns (they start with "ESG_score_")
esg_score_cols = [col for col in EU_ESG.columns if "ESG_score_" in col]

# Convert column names to years (e.g., 'ESG_score_2024' → 2024)
esg_years = [int(col.split("_")[-1]) for col in esg_score_cols]

# Calculate the average ESG score per year
avg_esg_per_year = EU_ESG[esg_score_cols].mean()

# Create a DataFrame for plotting
esg_trend = pd.DataFrame({"Year": esg_years, "Avg_ESG_Score": avg_esg_per_year.values})

# Sort by Year
esg_trend = esg_trend.sort_values("Year")

# Plot
plt.figure(figsize=(10, 5))
plt.plot(esg_trend["Year"], esg_trend["Avg_ESG_Score"], marker="o", linestyle="-", color="b")
plt.xlabel("Year")
plt.ylabel("Average ESG Score")
plt.title("Average ESG Score Over Time for European Companies")
plt.grid(True)

# plt.savefig("average_esg_score_EU.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
# Select only ESG score columns (they start with "ESG_score_")
esg_score_cols = [col for col in EU_ESG.columns if "ESG_score_" in col]

# Convert column names to years (e.g., 'ESG_score_2024' → 2024)
esg_years = [int(col.split("_")[-1]) for col in esg_score_cols]

# Count NaN values per year
nan_counts_per_year = EU_ESG[esg_score_cols].isna().sum()

# Create a DataFrame for plotting
nan_trend = pd.DataFrame({"Year": esg_years, "NaN_Count": nan_counts_per_year.values})

# Sort by Year
nan_trend = nan_trend.sort_values("Year")

# Plot
plt.figure(figsize=(10, 5))
plt.plot(nan_trend["Year"], nan_trend["NaN_Count"], marker="o", linestyle="-", color="r")
plt.xlabel("Year")
plt.ylabel("Number of NaN Values")
plt.title("Number of NaN ESG Scores Over Time for European Companies")
plt.grid(True)

# plt.savefig("nan_esg_score_EU.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
# Select only ESG score columns (they start with "ESG_score_")
esg_score_cols = [col for col in EU_ESG.columns if "ESG_score_" in col]

# Convert column names to years (e.g., 'ESG_score_2024' → 2024)
esg_years = [int(col.split("_")[-1]) for col in esg_score_cols]

# Calculate the percentage of NaN values per year
nan_percentage_per_year = EU_ESG[esg_score_cols].isna().mean() * 100  # Convert to percentage

# Create a DataFrame for plotting
nan_percentage_trend = pd.DataFrame({"Year": esg_years, "NaN_Percentage": nan_percentage_per_year.values})

# Sort by Year
nan_percentage_trend = nan_percentage_trend.sort_values("Year")

# Plot
plt.figure(figsize=(10, 5))
plt.plot(nan_percentage_trend["Year"], nan_percentage_trend["NaN_Percentage"], marker="o", linestyle="-", color="r")
plt.xlabel("Year")
plt.ylabel("Percentage of NaN Values (%)")
plt.title("Percentage of NaN ESG Scores Over Time for European Companies")
plt.grid(True)

plt.savefig("nan_percentage_esg_score_EU.png", dpi=300, bbox_inches="tight")

plt.show()
```
