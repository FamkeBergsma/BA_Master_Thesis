# Finance data USA

```python
data = pd.read_csv(r"file")
print(data.head())
```

```python
# Remove the first 10 rows of the dataset, since these contains only NaN's
data = data.iloc[4:].reset_index(drop=True)
# Remove the 'Size' column from the dataset

# Check the result
print(data.head(20))
```

```python
data.iloc[:, 3] #first column of dates for the ROE, last is at 27
data.iloc[:, 28] #first ROE, .last is at 52
```

```python
# Select company-related columns (Company Name and ISIN Code)
company_info_columns = ['Company Name', 'ISIN Code']
company_info = data[company_info_columns]

# Convert all ROE date columns (3 to 27) to datetime and extract years
for i in range(3, 28):
    data[f'Year_{i}'] = pd.to_datetime(data.iloc[:, i], format='%d/%m/%Y', errors='coerce').dt.year

# Initialize an empty DataFrame to store ROE per year
roe_per_year = pd.DataFrame(index=data.index)

# Loop through each row to correctly place ROE ratios in the right year columns
for index, row in data.iterrows():
    for i in range(3, 28):  # Date columns for ROE
        year = row[f'Year_{i}']  # Extract the year
        roe_column = data.columns[i + 25]  # ROE column (offset by 25)
        
        if pd.notna(year):  # If the year is valid
            roe_per_year.at[index, f"ROE_{int(year)}"] = row[roe_column]

# Sort columns from most recent year to oldest
roe_per_year = roe_per_year[sorted(roe_per_year.columns, key=lambda x: int(x.split('_')[-1]), reverse=True)]

# Concatenate company info with sorted ROE ratios
USA_finance = pd.concat([company_info, roe_per_year], axis=1)

# Display the resulting DataFrame
print(USA_finance)
```

```python
# Define a function to process market capitalization scores for USA_data
def process_market_cap_usa(USA_data, date_columns_range, market_cap_columns_offset):
    # Convert date columns to datetime and extract years
    for i in range(date_columns_range[0], date_columns_range[1] + 1):
        USA_data[f'Year_MarketCap_{i}'] = pd.to_datetime(USA_data.iloc[:, i], format='%d/%m/%Y', errors='coerce').dt.year

    # Create an empty DataFrame for market capitalization values
    market_cap_per_year = pd.DataFrame(index=USA_data.index)

    # Loop through each row to place market capitalization values in the corresponding year columns
    for index, row in USA_data.iterrows():
        for i in range(date_columns_range[0], date_columns_range[1] + 1):  # Date columns
            year = row[f'Year_MarketCap_{i}']  # Extract the year
            market_cap_column = USA_data.columns[i + market_cap_columns_offset]  # Market capitalization column (offset by 25)

            if pd.notna(year):  # If the year is valid
                market_cap_per_year.at[index, f"MarketCap_{int(year)}"] = row[market_cap_column]  # Ensure the year is an integer

    # Sort columns from most recent year to oldest
    market_cap_per_year = market_cap_per_year[sorted(market_cap_per_year.columns, key=lambda x: int(x.split('_')[-1]), reverse=True)]

    return market_cap_per_year

# Select company-related columns for USA_data
company_info_columns_usa = ['Company Name', 'ISIN Code', 'Ticker Symbol', 'Country of Exchange', 'NAICS National Industry Name']
company_info_usa = USA_data[company_info_columns_usa]

# Initialize an empty DataFrame for the final result
USA_control_variables = pd.DataFrame()

# Process Market Capitalization data and get values per year
market_cap_per_year_usa = process_market_cap_usa(USA_data, (406, 430), 25)

# Concatenate company info with the processed market capitalization data
USA_control_variables = pd.concat([company_info_usa, market_cap_per_year_usa], axis=1)

# Display the resulting DataFrame
print(USA_control_variables)
```

```python
# Export to Excel
USA_control_variables.to_excel('USA_control_variables.xlsx', index=False, engine='openpyxl')
```
