```python
# Define a function to process financial scores (ROE, ROA, EPS)
def process_financial_scores(EU_data, score_type, date_columns_range, score_columns_offset):
    # Convert date columns to datetime and extract years
    for i in range(date_columns_range[0], date_columns_range[1] + 1):
        EU_data[f'Year_{score_type}_{i}'] = pd.to_datetime(EU_data.iloc[:, i], format='%d/%m/%Y', errors='coerce').dt.year

    # Create an empty DataFrame for the scores
    scores_per_year = pd.DataFrame(index=EU_data.index)

    # Loop through each row to place scores in the corresponding year columns
    for index, row in EU_data.iterrows():
        for i in range(date_columns_range[0], date_columns_range[1] + 1):  # Date columns
            year_score = row[f'Year_{score_type}_{i}']  # Extract the year
            score_column = EU_data.columns[i + score_columns_offset]  # Score column (offset by 25)
            
            if pd.notna(year_score):  # If the year is valid
                scores_per_year.at[index, f"{score_type}_score_{year_score}"] = row[score_column]

    # Sort columns from most recent year to oldest
    scores_per_year = scores_per_year[sorted(scores_per_year.columns, key=lambda x: int(x.split('_')[-1]), reverse=True)]

    return scores_per_year

# Select company-related columns
company_info_columns = ['Company Name', 'ISIN Code', 'Ticker Symbol', 'Country of Exchange', 'NAICS National Industry Name']
company_info = EU_data[company_info_columns]

# Initialize an empty DataFrame for the final result
finance_performance_by_year = pd.DataFrame()

# Process ROE, ROA, and EPS data and get scores per year
roe_scores_per_year = process_financial_scores(EU_data, 'ROE', (456, 480), 25)
roa_scores_per_year = process_financial_scores(EU_data, 'ROA', (506, 530), 25)
eps_scores_per_year = process_financial_scores(EU_data, 'EPS', (556, 580), 25)

# Concatenate company info with the processed scores
finance_performance_by_year = pd.concat([company_info, roe_scores_per_year, roa_scores_per_year, eps_scores_per_year], axis=1)

# Display the resulting DataFrame
print(finance_performance_by_year)

```

```python
# Define a function to process market capitalization scores
def process_market_cap(data, date_columns_range, market_cap_columns_offset):
    # Convert date columns to datetime and extract years
    for i in range(date_columns_range[0], date_columns_range[1] + 1):
        data[f'Year_MarketCap_{i}'] = pd.to_datetime(data.iloc[:, i], format='%d/%m/%Y', errors='coerce').dt.year

    # Create an empty DataFrame for market capitalization values
    market_cap_per_year = pd.DataFrame(index=data.index)

    # Loop through each row to place market capitalization values in the corresponding year columns
    for index, row in data.iterrows():
        for i in range(date_columns_range[0], date_columns_range[1] + 1):  # Date columns
            year = row[f'Year_MarketCap_{i}']  # Extract the year
            market_cap_column = data.columns[i + market_cap_columns_offset]  # Market capitalization column (offset by 25)

            if pd.notna(year):  # If the year is valid
                market_cap_per_year.at[index, f"MarketCap_{int(year)}"] = row[market_cap_column]  # Ensure the year is an integer

    # Sort columns from most recent year to oldest
    market_cap_per_year = market_cap_per_year[sorted(market_cap_per_year.columns, key=lambda x: int(x.split('_')[-1]), reverse=True)]

    return market_cap_per_year

# Select company-related columns
company_info_columns = ['Company Name', 'ISIN Code', 'Ticker Symbol', 'Country of Exchange', 'NAICS National Industry Name']
company_info = data[company_info_columns]

# Initialize an empty DataFrame for the final result
EU_control_variables = pd.DataFrame()

# Process Market Capitalization data and get values per year
market_cap_per_year = process_market_cap(data, (406, 430), 25)

# Concatenate company info with the processed market capitalization data
EU_control_variables = pd.concat([company_info, market_cap_per_year], axis=1)

# Display the resulting DataFrame
print(EU_control_variables)
```
